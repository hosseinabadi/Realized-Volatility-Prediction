```python
import pandas as pd
```

```python
path = './sp500_news_and_sentimental_weights.csv'
df = pd.read_csv(path)
df.drop(columns=['Unnamed: 0'], inplace=True)
df.head()
```

```python
# compute weighted sentiments
df['positivity_text_weighted'] = df['positivity_text'] * (1+df['weight'])
df['negativity_text_weighted'] = df['negativity_text'] * (1+df['weight'])
df['neutrality_text_weighted'] = df['neutrality_text'] * (1+df['weight'])
```

```python
df.to_csv('sp500_weighted_sentiments.csv')
```
