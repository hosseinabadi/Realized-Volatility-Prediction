```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot_results(predicted, title, directory):
    plt.figure(figsize=(12, 6))
    plt.plot(predicted.index, predicted['Realized Volatility'], label='Actual', color='black', linewidth=2)
    plt.plot(predicted.index, predicted['Predicted Volatility'], label=f'{title} Prediction', color='blue', linestyle='--')
    # plt.title(f'{title} Prediction vs Actual', fontsize=16)
    plt.xlabel('Time', fontsize=14)
    plt.ylabel('Realized Volatility', fontsize=14)
    plt.legend()
    plt.tight_layout()
    plt.savefig(f"Outputs/Models/{directory}/{title}_prediction_vs_actual.png")
    plt.show()
```

```python
'''
Plot the results of GARCH and LSTM with single input
'''
garch = pd.read_csv("Outputs/Models/GARCH/GARCH_results.csv", index_col="Date", parse_dates=True)
LSTM = pd.read_csv("Outputs/Models/LSTM_single_input/LSTM_single_input_RV_results.csv", index_col="Unnamed: 0", parse_dates=True)
garch = garch.loc[LSTM.index]

plot_results(garch, "GARCH", "GARCH")
plot_results(LSTM, "LSTM", "LSTM_single_input")
```

```python
'''
Plot the results of VAR model
'''
VAR = pd.read_csv("Outputs/Models/VAR/VAR_results.csv", index_col="Date", parse_dates=True)
plot_results(VAR, "VAR", "VAR")
```

```python
'''
Plot the results of LSTM model with multiple inputs
'''
LSTM = pd.read_csv("Outputs/Models/LSTM_multiple_input/LSTM_results.csv", index_col="Unnamed: 0", parse_dates=True)
plot_results(LSTM, "LSTM","LSTM_multiple_input")
```

```python
'''
Plot the results of XGBoost model
'''
XG = pd.read_csv("Outputs/Models/XGBoost/XGBoost_results.csv", index_col="Unnamed: 0", parse_dates=True)
plot_results(XG, "XGBoost", "XGBoost")
```

```python
'''
Plot the results of Random Forest model
'''
RF = pd.read_csv("Outputs/Models/Random Forest/RandomForest_results.csv", index_col="Unnamed: 0", parse_dates=True)
plot_results(RF, "Random Forest", "Random Forest")
```

```python
'''
Plot the results of Recurrent Neural Network model
'''

RNN = pd.read_csv("Outputs\Models\RNN\RNN_results.csv", index_col="Unnamed: 0", parse_dates=True)
plot_results(RNN, "RNN", "RNN")
```

```python
'''
Plot the results of linear regression model
'''
LR = pd.read_csv("Outputs/Models/Linear Regression/linear_regression_results.csv", index_col="Unnamed: 0", parse_dates=True)
plot_results(LR, "Linear Regression", "Linear Regression")
```

```python
'''
Plot the results of Lasso model
'''
Lasso = pd.read_csv("Outputs/Models/Lasso/Lasso_results.csv", index_col="Unnamed: 0", parse_dates=True)
plot_results(Lasso, "Lasso", "Lasso")
```

```python
'''
Plot the results of Ridge model
'''
Ridge = pd.read_csv("Outputs/Models/Ridge/Ridge_results.csv", index_col="Unnamed: 0", parse_dates=True)
plot_results(Ridge, "Ridge", "Ridge")
```

```python
def plot_combined_results(data, labels, title):
    plt.figure(figsize=(12, 6))
    for i in range(len(data)):
        plt.plot(data[i].index, data[i]['Predicted Volatility'], label=labels[i], linestyle='--')

    plt.plot(LR.index, LR['Realized Volatility'], label='Actual', color='black', linewidth=2)

    plt.xlabel('Date', fontsize=14)
    plt.ylabel('Realized Volatility', fontsize=14)
    # plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.savefig(f"Outputs/Models/{title}.png")
    plt.show()

# Plot the combined results
plot_combined_results([LR, Lasso, Ridge],["Linear Regression", "Lasso", "Ridge"], "Linear Regression, Lasso, and Ridge")
```

```python
plot_combined_results([LSTM, RNN],["LSTM", "RNN"], "LSTM and RNN")
```

```python
plot_combined_results([XG, RF],["XGBoost", "Random Forest"], "XGBoost and Random Forest")
```
